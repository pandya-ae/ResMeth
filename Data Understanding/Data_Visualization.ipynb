##Pie Chart

```python
flavors = ('Chocolate', 'Vanilla', 'Pistachio', 'Mango', 'Strawberry')
votes = (12, 11, 4, 8, 7)
colors = ('#8B4513', '#FFF8DC', '#93C572', '#E67F0D', '#D53032')
explode = (0, 0, 0, 0.1, 0)

plt.title('Favorite Ice Cream Flavors')
plt.pie(
    votes,
    labels=flavors,
    autopct='%1.1f%%',
    colors=colors,
    explode=explode,
    shadow=True
    )
plt.show()
```

## Grouping

```python
df['drive-wheels'].unique()
```

```python
df_group = df[['drive-wheels','body-style','price']]
df_group
```

```python
# grouping results
df_group_one = df_group.groupby(['drive-wheels'],as_index=False).mean()
df_group_one
```

```python
# grouping results
df_group_two = df_group.groupby(['drive-wheels','body-style'],as_index=False).mean()
df_group_two
```

```python
# Pivot table
grouped_pivot = df_group_two.pivot(index='drive-wheels',columns='body-style')
grouped_pivot = grouped_pivot.fillna(0) #fill missing values with 0
grouped_pivot
```

```python
# Use the "groupby" function to find the average "price" of each car by "body-style" ?
df_gptest2 = df[['body-style','price']]
grouped_test_bodystyle = df_gptest2.groupby(['body-style'],as_index= False).mean()
grouped_test_bodystyle
```

```python
#Create heatmap using the grouped_pivot results
plt.pcolor(grouped_pivot, cmap='RdBu')
plt.colorbar()
plt.show()
```

```python
fig, ax = plt.subplots()
im = ax.pcolor(grouped_pivot, cmap='RdBu')
 
#label names
row_labels = grouped_pivot.columns.levels[1]
col_labels = grouped_pivot.index
 
#move ticks and labels to the center
ax.set_xticks(np.arange(grouped_pivot.shape[1]) + 0.5, minor=False)
ax.set_yticks(np.arange(grouped_pivot.shape[0]) + 0.5, minor=False)
 
#insert labels
ax.set_xticklabels(row_labels, minor=False)
ax.set_yticklabels(col_labels, minor=False)
 
#rotate label if too long
plt.xticks(rotation=90)
 
fig.colorbar(im)
plt.show()
```
